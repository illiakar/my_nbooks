```python
import pandas as pd
import seaborn as sns
```

```python
avocadoURL = 'avocado.csv'
avocadoDf = pd.read_csv(avocadoURL)
avocadoDf.head()
```

```python
avocadoDf.info()
```

```python
avocadoDf.Date = avocadoDf.Date.apply(pd.to_datetime)
avocadoDf.info()
```

```python
avocadoSubDf = avocadoDf.query("year == 2015 and type == 'conventional' and region == 'Albany'")[['Date', 'Total Bags', 'Small Bags']]
avocadoSubDf.head()
```

```python
avocadoDfMelted = pd.melt(avocadoDf, id_vars='Date', value_vars=['Total Bags','Small Bags'], var_name='Bags', value_name='Count')
avocadoDfMelted
```

```python
sns.relplot(data=avocadoDfMelted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocadoDf['Sales Volume'] = pd.qcut(avocadoDf['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
avocadoDf.head()
```

```python
sns.countplot(data=avocadoDf, x='year', hue='Sales Volume')
```

```python
avocadoDfMelted2 = pd.melt(avocadoDf, id_vars='year', value_vars=['Small Bags','Large Bags','XLarge Bags'], var_name='Bags', value_name='Count')
avocadoDfMelted2
```

```python
sns.relplot(data=avocadoDfMelted2, kind='line', x='year', y='Count', hue='Bags')
```
